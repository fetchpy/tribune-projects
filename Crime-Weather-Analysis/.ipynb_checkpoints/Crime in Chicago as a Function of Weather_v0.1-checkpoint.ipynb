# <center> The effect of weather temperature on frequency of shootings </center>

In this notebook, I examine the relationship between weather temperature and frequency of shootings. This analysis is based on shootings data from **January,1st 2012** until **June, 7th 2017**. This dataset was pulled from City of Chicago Data Portal [link](https://data.cityofchicago.org/Public-Safety/Crimes-One-year-prior-to-present/x2n5-8w5q). I also collected historical weather data from **January,1st 2012** until **June, 7th 2017** [link](http://mrcc.isws.illinois.edu/CLIMATE/restricted.jsp?accessdenied=%2FCLIMATE%2Fwelcome.jsp).

```python
#importing libraries
import pandas as pd
import numpy as np
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import glob
%matplotlib inline
```

```python
#Compiling multiple csv files present in a folder together in one dataframe.
data = pd.DataFrame()
for f in glob.glob('./crimeData/*.csv'): # all files in the directory that matchs the criteria.
    data = pd.concat([data,pd.read_csv(f)])
```

```python
data.shape
```

```python
data.columns
```

```python
#Reading in Weather Data

temp = pd.read_csv('./climate_Data_Jan_1_2012_up_to_Jun_7_2017.csv')
temp.columns
```

```python
'''
cleanDate_W: cleans and re-formats the date column in the weather data
'''
def cleanDate_W(d):
    return (datetime.strptime(str(d), '%m/%d/%y').date())
temp['Date'] = temp['Date'].apply(cleanDate_W)
temp['Date']= pd.to_datetime(temp['Date'],errors='coerce')
```

```python
'''
cleanDate_S: cleans and re-formats the date column in the shootings data
'''
def cleanDate_S(d):
    if(d==''):
        return (d)
    else:
        return (datetime.strptime(str(d) ,'%m/%d/%Y %H:%M:%S %p').date()) #converting dates from type string to type date and time
data['Date'] = data['Date'].apply(cleanDate_S)
data['Date'] = pd.to_datetime(data['Date'],errors='coerce')
```

```python
#merging the shootings and temp data based on the Date column
crimeTemp = data.merge(temp,how='inner',left_on='Date',right_on='Date')
```

```python
crimeTemp.shape
```

```python
#checking if the new column (Mean_Temp) has been added
crimeTemp.columns
```

```python
crimeTemp.head(4)
```

```python
crimeCount = pd.DataFrame({'S_Date':crimeTemp['Date'].tolist(),
                          'Temp':crimeTemp['Mean_Temp'].tolist(),
                          'Crime':crimeTemp['Primary Type'].tolist()})
```

```python
'''
C_count: is a dataframe that includes the number of crimes per day as well as the average temperature of that day.
'''
C_count = pd.DataFrame(crimeCount.groupby(['S_Date','Crime','Temp'])['Temp'].count())
C_count.columns= ['Number of Crimes']
C_count = C_count.reset_index()
C_count
```

```python
'''
Visualizing the crime patterns as a function of temperature.
'''
plt.figure(figsize=(15,15))
sns.lmplot('Temp','Number of Crimes',hue='Crime',data=C_count,scatter=False)
plt.title('Crime and Temperature',fontsize=15)
```

```python
'''
breaking down the plot by crime type.
Summary: Battery and Theft tend to have the steepest slopes, or highest trend, in explaining the relationship 
between crime and temperature. In other words, as temperature increases the the number of battery and theft increases too.
'''
plt.figure(figsize=(15,15))
sns.lmplot('Temp','Number of Crimes',hue='Crime',data=C_count,scatter=False,col='Crime',col_wrap=3)
```

```python
C_count['Year'] = C_count['S_Date'].dt.year
```

```python
'''
Breaking down the results by year.
'''
plt.figure(figsize=(15,15))
sns.lmplot('Temp','Number of Crimes',hue='Crime',data=C_count,scatter=False,col='Year',col_wrap=3)
```

```python
'''
Checking the general trend, or relationship, between the number of crimes and temperature.
The confidence interval is very small, indicating higher accuracy.
'''
plt.figure(figsize=(15,15))
sns.regplot('Temp','Number of Crimes',data=C_count,scatter=False)
```

```python
'''
Visualizing data patterns/linear distribution
'''
plt.figure(figsize=(15,15))
sns.lmplot('Temp','Number of Crimes',hue='Crime',data=C_count,scatter=True,col='Year',col_wrap=3,fit_reg=False)
#plt.title('Crime and Temperature',fontsize=15)
```

```python
import statsmodels.api as sm
```

```python
import statsmodels.formula.api as smf
```

```python
'''
Run a regression model to predict number of shootings from temperature
'''
regData = C_count
regData.columns=['S_Date', 'Crime', 'Temp', 'Num', 'Year']
results = smf.ols("Num ~ Temp", data=regData).fit()
```

```python
print(results.summary())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
'''
Train a machine learning regression model to predict the data
'''
X = C_count[['Num', 'Year']]
y = C_count['Temp']
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
lm= LinearRegression()
lm.fit(x_train,y_train)
```

```python
predictions= lm.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
'''
The amount of variance expalined by the model
'''
r2_score(y_test,predictions)
```

```python
'''
Convert dates (%Y-%m-%d) to days (Sunday, Monday...etc.)
'''
def toDay(d):
    return (datetime.strptime(str(d),'%Y-%m-%d %H:%M:%S').date().strftime('%A'))
C_count['Day'] = C_count['S_Date'].apply(toDay)
```

```python
dictDay= {'Saturday':0,'Sunday':0,'Friday':0,'Monday':1,'Tuesday':1,'Wednesday':1,'Thursday':1,'Friday':1}
C_count['Day']= C_count['Day'].map(dictDay)
```

```python
'''
Trying to find if the relationship between crime and temperature changes between weekends and weekedays.
Summary: Interestingly, only Battery and Criminal Damage tend to be high on weekends as the temperature increases.
Some of the remaining types of crimes, however, show an increase on weekdays, but not on weekends.
'''
plt.figure(figsize=(15,15))
sns.lmplot('Temp','Num',hue='Day',data=C_count,scatter=False,col='Crime',col_wrap=3)
#summary: 
#results below show that only batter tend to occure more frequently on warm days especially on weekends (which is similar to the shootings pattern).
#However, remaining patterns are mainly happening on weekdays
```

# <center> Statistical Analysis</center>

```python
'''
Reformating data to control for the type of crime in the model.
The following line of code reformats C_count['Crime'].to_dict to {'Crime_Name':12345}
'''
crime_Codes = {y:x for x,y in C_count['Crime'].to_dict().items()}
```

```python
'''
mapping each crime name to its number code
'''
C_count ['Crime_Code'] = C_count['Crime'].map(crime_Codes)
```

```python
X = C_count[['Crime_Code','Num','Year']]
y = C_count['Temp']
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.3, random_state=42)
lm= LinearRegression()
lm.fit(x_train,y_train)
predictions= lm.predict(x_test)
```

```python
r2_score(y_test,predictions)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
#The smaller the number is the better. This measures how far are our predictions from the original/real values
mean_absolute_error(y_test,predictions)
```

```python
'''
Taking the square root of the mean squared error converts the units back to the original 
units of the output variable and can be meaningful for description and presentation
source: http://machinelearningmastery.com/metrics-evaluate-machine-learning-algorithms-python/
'''
np.sqrt(mean_squared_error(y_test,predictions))
```
